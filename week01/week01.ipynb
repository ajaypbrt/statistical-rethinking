```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Problem 1

```python
N = 101
p_values = np.linspace(0, 1, N)
prior = [1] * N
posterior = np.array([prior[i] * binom.pmf(4, 15, p_values[i]) for i in range(N)])
posterior = posterior / posterior.sum()
plt.plot(p_values, posterior)
```

### Problem 2

```python
N = 101
p_values = np.linspace(0, 1, N)
prior = [0 if p_values[i] < 0.5 else 1 for i in range(N)]
posterior = np.array([prior[i] * binom.pmf(4, 6, p_values[i]) for i in range(N)])
posterior = posterior / posterior.sum()
plt.plot(p_values, posterior)
```
